This notebook demonstrates the basic concepts of **quantum states** for a single qubit.

## ‚öõÔ∏è Qubit Theory

A **qubit** (quantum bit) is the basic unit of quantum information, analogous to a bit in classical computing.

### Basis States:
The two fundamental quantum states are represented as:

\[
|0‚ü© = [1,0] , 
|1‚ü© = [0,1] 
\]

These are known as the **computational basis states**.

### Superposition:
A qubit can exist in a *superposition* of |0‚ü© and |1‚ü©:

\[
|œà‚ü© = Œ±|0‚ü© + Œ≤|1‚ü©
\]

where Œ± and Œ≤ are complex numbers satisfying:

\[
|Œ±|^2 + |Œ≤|^2 = 1
\]

---


## üíª Implementation

We will now create the basic |0‚ü© and |1‚ü© qubit states in Python using **NumPy** arrays.


```python
import numpy as np

# Define basis states
zero_state = np.array([[1], [0]])  # |0‚ü©
one_state = np.array([[0], [1]])   # |1‚ü©

print("Qubit |0‚ü© state:")
print(zero_state)

print("\nQubit |1‚ü© state:")
print(one_state)
```

## üåê Bloch Sphere Visualization

We can represent qubit states on the **Bloch sphere**, a 3D visualization of the qubit‚Äôs state vector.

Below, we show the |0‚ü© and |1‚ü© states as points on the Bloch sphere.


```python
from qutip import Bloch
import numpy as np

# Create the Bloch sphere
b = Bloch()

# Define |0‚ü© and |1‚ü© state vectors
zero = np.array([[1], [0]])
one = np.array([[0], [1]])

# Add vectors for |0‚ü© and |1‚ü©
b.add_vectors([0, 0, 1])   # |0‚ü©
b.add_vectors([0, 0, -1])  # |1‚ü©

# Customize and show
b.vector_color = ['blue', 'red']
b.point_color = ['blue', 'red']
b.xlabel = ['$x$', '']
b.ylabel = ['$y$', '']
b.zlabel = ['$|0‚ü©$', '$|1‚ü©$']
b.show()
```
