# Imports for the Single Qubit Gates 

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
import matplotlib.pyplot as plt
```

# Visualization Function so that gates can be called Recursively 

### takes circuit(Gate) and Title and Visualize it 

```python
def visualize_gate(circuit, title):
    state = Statevector.from_instruction(circuit)
    print(f"\n{title}")
    display(circuit.draw('mpl'))
    display(plot_bloch_multivector(state))
```

## Test Case 1: Initial |0⟩ State

# Initial state of QUBIT

```python
qc = QuantumCircuit(1)
visualize_gate(qc, "Initial State |0⟩")
```

# Applying Hardmard Gate 

### hardmard gate puts the qubit back into superpositon state where it get 50% chances to be both the states

```python
qc_h = QuantumCircuit(1)
qc_h.h(0)
visualize_gate(qc_h, "After Hadamard (H) Gate")
```

# Applying the Pauli - x gate 

### it performs a reflection around the x axis its like a  bit flip which swaps the amplitude 

```python
qc_x = QuantumCircuit(1)
qc_x.x(0)
visualize_gate(qc_x, "After Pauli-X Gate")
```

# applying pauli - Y gate 

### phase factor + pauli X gate

```python
qc_y = QuantumCircuit(1)
qc_y.y(0)
visualize_gate(qc_y, "After Pauli-Y Gate")
```

# Applying Pauli - Z gate 

### phase flip leave the satate zero unchnaged 

```python
qc_z = QuantumCircuit(1)
qc_z.z(0)
visualize_gate(qc_z, "After Pauli-Z Gate")
```

# Applying phasse gate

### leave zero unchaged but multiply 1 by i 

```python
qc_s = QuantumCircuit(1)
qc_s.s(0)
visualize_gate(qc_s, "After Phase (S) Gate")
```

## Test Case 2: Gates on Superposition

```python
qc_hx = QuantumCircuit(1)
qc_hx.h(0)
qc_hx.x(0)
visualize_gate(qc_hx, "Hadamard → Pauli-X")
```

```python
qc_hz = QuantumCircuit(1)
qc_hz.h(0)
qc_hz.z(0)
visualize_gate(qc_hz, "Hadamard → Pauli-Z")
```

```python
qc_hs = QuantumCircuit(1)
qc_hs.h(0)
qc_hs.s(0)
visualize_gate(qc_hs, "Hadamard → Phase (S) Gate")
```
